```python
import matplotlib.pyplot as plt
x=["python","c","cpp","java","R"]
y=[82,30,90,65,50]
z=[94,56,74,23,77]

plt.xlabel("launguage",fontsize=15)
plt.ylabel("Value",fontsize=15)
plt.title("Programming lag.",fontsize=15)

# width ::
c=["y","r","g","b","m"]
# if we want to change the align you can by align : 1.edge 2.center
# we can also change boundary color called as edge color
# line width for edge size
# line style chacnge by eg. : this for dot line  # linestyle=":"
# clor light by alpha parameter
# label use for label and plt .legend
plt.bar(x,y,width=0.5,color="red",align="edge",edgecolor="black",linewidth=2,alpha=0.9,label="popularity")

# multiple bar graph under single one
plt.bar(x,z,width=0.4,color="blue",edgecolor="blue",linewidth=2,label="popularity")

# x axis name:
# rote name
plt.xticks(rotation=10)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x=["python","c","cpp","java","R"]
y=[82,30,90,65,50]


plt.xlabel("launguage",fontsize=15)
plt.ylabel("Value",fontsize=15)
plt.title("Programming lag.",fontsize=15)


plt.barh(x,y,color="red",align="edge",height=0.4)

# multiple bar graph under single one
plt.barh(x,y,height=0.3,align="center",color="yellow")
plt.barh(x,y,height=0.1,align="center",color="blue")

plt.show()
```

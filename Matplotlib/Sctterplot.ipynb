```python
import matplotlib.pyplot as plt
x=["Mon","Tue","Wed","Thr","Fri","Sat","Sun"]
y=[90,57,88,74,32,45,100]
z=[222,74,97,23,90,87,19]

# we can change every point size::
# we can chacge transperency
# we change marker :  we change marker boundry color elc
sizes=[100,600,500,800,300,700,606]

# we can also have color map in which we give number
# cmap having diffrent color names
colors=[10,47,33,78,32,89,55] #between10 - 100
plt.scatter(x,y,c=colors,s=sizes,alpha=0.9,marker="|",edgecolor="black",linewidth=1,cmap="PuRd")
plt.scatter(x,z,c=colors,s=sizes,cmap="CMRmap_r")
plt.title("Scatter PLot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("Work",fontsize=15)

# we can also able to show colro bar:
t=plt.colorbar()
t.set_label("ColorBar",fontsize=15)
plt.show()

```

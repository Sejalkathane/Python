# PiePlot

```python
import matplotlib.pyplot as plt
x=[57,23,76,72,11]
y=["C","CPP","JAVA","PYTHON","REACT"]
ex=[0.0,0.0,0.0,0.0,0.3]
# using explode seperate small part
c=["red","blue","green","yellow","m"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.2f%%",shadow=True,radius=1.5,labeldistance=1.1,startangle=0
       ,textprops={"fontsize": 12},counterclock=False,wedgeprops={"linewidth":5,"edgecolor":"white"},rotatelabels=True)
plt.title("Alpa")
plt.legend(loc=2)
plt.show()

```

```python
# dot pie chat
plt.pie([1])
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[57,23,76,72,11]
x1=[67,38,84,25,90]
y=["C","CPP","JAVA","PYTHON","REACT"]
c=["r","g","y","b","m"]
plt.pie(x,labels=y,radius=2)
plt.pie(x1,radius=1.2,colors=c)
# piechar inside piechart
plt.show()
```
